```python
 # Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

## __District Summary__

```python
school_data.head(5)
```

```python
student_data.head ()
```

```python
# Calculate the total number of schools

total_schools = school_data["school_name"].nunique()
total_schools


# Calculate the percentage of students with a passing math score (70 or greater)
# Calculate the percentage of students with a passing reading score (70 or greater)
# Calculate the percentage of students who passed math and reading (% Overall Passing)
# Create a dataframe to hold the above results
# Optional: give the displayed data cleaner formatting
```

```python
# Calculate the total number of students
total_numer_of_students = student_data ["Student ID"].count()
total_numer_of_students
```

```python
# Calculate the total budget
total_budget= school_data["budget"].sum ()
total_budget
```

```python
# Calculate the average math score 
average_math_score = student_data["math_score"].mean ()
average_math_score
```

```python
# Calculate the average reading score
average_reading_score = student_data["reading_score"].mean ()
average_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
pass_math = student_data[student_data["math_score"]>=70].count()["student_name"]
percentage_passed_math = ((pass_math/total_numer_of_students)*100)
percentage_passed_math
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading = student_data[student_data["reading_score"]>=70].count()["student_name"]
percentage_passed_reading = ((pass_reading/total_numer_of_students)*100)
percentage_passed_reading
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
percentage_pass_combined = student_data[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70) ].count()["student_name"]
percentage_rate_combined = percentage_pass_combined/total_numer_of_students*100
percentage_rate_combined
```

```python
# Create a dataframe to hold the above results

District_Summary_df = pd.DataFrame({"Total Schools":[total_schools],
                      "Total Students":[total_numer_of_students],
                      "Total Budget":[total_budget],
                      "Average Math Score":[average_math_score],
                      'Average Reading Score':[average_reading_score],
                      "% Passing Math":[percentage_passed_math],
                      "% Passing Reading":[percentage_passed_reading],
                      "Overall Passing rate":[percentage_pass_combined]})

District_Summary_df
```

```python
# Combine the data into a single dataset.  
school_students_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_students_df.head()
```

## __School Summary__

```python
#Create an overview table that summarizes key metrics about each school, including:
#School Name
#school_students_df["school_name"].unique()
#School Type
#Total Students
#Total School Budget
#Per Student Budget
#Average Math Score
#Average Reading Score
#% Passing Math
#% Passing Reading
#% Overall Passing (The percentage of students that passed math and reading.)
```

```python
#School Type
school_type = school_data.set_index(["school_name"])["type"]
school_type.head
```

```python
#Total Students
grouped_school_students_df = school_students_df.groupby("school_name")
total_students_school= grouped_school_students_df.count()["student_name"]
total_students_school
```

```python
#Total School Budget
total_budget= grouped_school_students_df.mean()["budget"]
total_budget
```

```python
#Per Student Budget
per_student_budget= (total_budget/total_students_school)
per_student_budget.round(3)
```

```python
#Average Math Score
avg_math_school= grouped_school_students_df.mean()["math_score"]
avg_math_school
```

```python
#Average Reading Score
avg_reading = grouped_school_students_df.mean()["reading_score"]
avg_reading
```

```python
math_df=grouped_school_students_df["math_score"].head()
```

```python
mathList=math_df.values.tolist()
```

```python
#% Passing math
pass_math_school = school_students_df[school_students_df["math_score"] >= 70] 
```

```python
per_school_passing_math = pass_math_school.groupby(["school_name"]).count()["student_name"]/ total_students_school * 100
per_school_passing_math
```

```python
#% Passing Reading
pass_reading_school = school_students_df[(school_students_df["reading_score"]>=70)]
per_school_passing_reading = pass_reading_school.groupby(["school_name"]).count()["student_name"]/ total_students_school * 100
per_school_passing_reading
```

```python
#% Overall Passing (The percentage of students that passed math and reading.)
percentage_pass_combined = student_data[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)]
percentage_rate_combined = percentage_pass_combined.groupby(["school_name"]).count()["student_name"] /total_students_school*100
percentage_rate_combined
```

```python
#Create a dataframe to hold the above results
# School_Summary_df.set_index(school_name)
School_Summary_df = pd.DataFrame({"School Type":school_type,
                      "Total Students":total_students_school,
                      "Total Budget":total_budget,
                      "Per Student Budget":per_student_budget,
                      "Average Math Score":avg_math_school,
                      'Average Reading Score':avg_reading,
                      "% Passing Math":per_school_passing_math,
                      "% Passing Reading":per_school_passing_reading,
                      "Overall Passing rate":percentage_rate_combined})
# School_Summary_df[School_Summary_df.index.duplicated()]                    

School_Summary_df
```

## __Top Performing Schools (By % Overall Passing)__

## __Math Scores by Grade__

```python
#Math Scores by Grade
ninth_graders = school_students_df[(school_students_df["grade"]== "9th")]
tenth_graders = school_students_df[(school_students_df["grade"]== "10th")]
eleventh_graders = school_students_df[(school_students_df["grade"]== "11th")]
twelfth_graders = school_students_df[(school_students_df["grade"]== "12th")]
ninth_graders_scores= ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]



scores_by_grade = pd.DataFrame ({ "Ninth graders": ninth_graders_scores,
                                 "Tenth graders": tenth_graders_scores,
                                 "Eleventh graders": eleventh_graders_scores,
                                 "Twelfth graders": twelfth_graders_scores
})

scores_by_grade
```

## __Reading Scores by Grade__

```python
#Reading Scores by Grade
ninth_graders_r = school_students_df[(school_students_df["grade"]== "9th")]
tenth_graders_r = school_students_df[(school_students_df["grade"]== "10th")]
eleventh_graders_r = school_students_df[(school_students_df["grade"]== "11th")]
twelfth_graders_r = school_students_df[(school_students_df["grade"]== "12th")]
ninth_graders_scores_r= ninth_graders_r.groupby(["school_name"]).mean()["reading_score"]
tenth_graders_scores_r = tenth_graders_r.groupby(["school_name"]).mean()["reading_score"]
eleventh_graders_scores_r = eleventh_graders_r.groupby(["school_name"]).mean()["reading_score"]
twelfth_graders_scores_r = twelfth_graders_r.groupby(["school_name"]).mean()["reading_score"]



scores_by_grade_r = pd.DataFrame ({ "Ninth graders": ninth_graders_scores_r,
                                 "Tenth graders": tenth_graders_scores_r,
                                 "Eleventh graders": eleventh_graders_scores_r,
                                 "Twelfth graders": twelfth_graders_scores_r
})

scores_by_grade_r
```

```python
#Create bins

spending_bins = [0, 585, 630, 645, 676]
spending_names = ["<$584", "$585-629", "$630-644", "$645-675"]

School_Summary_df["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=spending_names)
School_Summary_df
```

```python
Size_math_score = School_Summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
Size_reading_score = School_Summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
Size_math_passing = School_Summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
Size_reading_passing = School_Summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
Size_overall_passing = School_Summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Overall Passing rate"]
```

## __Scores by School Spending__

```python
Scores_by_School = pd.DataFrame ({ "Average Math Score":Size_math_score,
                        "Average Reading Score":Size_reading_score,
                        "% Passing Math":Size_math_passing,
                        "% Passing Reading":Size_reading_passing,
                        "Overall Passing Rate":Size_overall_passing
})
Scores_by_School
```

## __Scores by School Spending__

```python
#Scores by school size
#Create bins

school_size_bins = [0, 1000, 2000, 5001]
school_size_names = ["Small (<1000) ", "Medium (1000-2000)", "Large (2000-5000)"]

School_Summary_df["School Size"] = pd.cut(School_Summary_df["Total Students"], school_size_bins, labels=school_size_names)
School_Summary_df
```

```python
Scores_School_Size = pd.DataFrame ({ "Average Math Score":Size_math_score_school,
                        "Average Reading Score":Size_reading_score_school,
                        "% Passing Math":Size_math_passing_school,
                        "% Passing Reading":Size_reading_passing_school,
                        "Overall Passing Rate":Size_overall_passing_school
})
Scores_School_Size
```
